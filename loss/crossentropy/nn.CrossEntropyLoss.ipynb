```
import torch
import torch.nn as nn
```

```
sequence_length = 1
batch_size = 16
input_size = 30
hidden_size = 32
num_layers = 2
output_size = 1
```

```
input = torch.randn(sequence_length, batch_size, input_size, requires_grad=True)
target = torch.empty(batch_size, output_size, dtype=torch.long).random_(2)
```

```
model = nn.LSTM(input_size, hidden_size, num_layers)
h0 = torch.randn(num_layers, batch_size, hidden_size)
c0 = torch.randn(num_layers, batch_size, hidden_size)
fc = nn.Linear(hidden_size, output_size, bias = True)
```
