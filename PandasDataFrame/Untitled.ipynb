```python
from sklearn.preprocessing import MinMaxScaler ; scaler = MinMaxScaler()
import pandas as pd
```

```python
gapminder = pd.read_csv('./data/gapminder.tsv', sep='\t')
```

```python
gapminder.head()
```

```python
lst = ['year', 'country', 'pop', 'gdpPercap']
AsiaDataSet = gapminder[gapminder['continent'] == 'Asia']
df = AsiaDataSet[lst].reset_index(drop=True)
```

```python
df.head()
```

```python
des_col_1 = ['year', 'country']
des_col_2 = ['pop', 'gdpPercap']
o_df = df[des_col_1]
n_df = df[des_col_2]
```

```python
o_df.head()
```

```python
n_df.head()
```

```python
scaled_data_np = scaler.fit_transform(n_df) # scaling을 거치면 np.array 형태로 변환됨
```

```python
scaled_df = pd.DataFrame(scaled_data_np)
scaled_df.columns = ['pop', 'gdpPercap']
```

```python
scaled_df
```

```python
new_df = pd.concat([o_df, scaled_df], axis=1)
```
